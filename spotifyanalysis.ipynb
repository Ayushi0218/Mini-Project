```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
tracks = pd.read_csv('Dataset/Spotify Dataset.csv')
```

```python
tracks.head()
```

```python
tracks.isnull().sum()
```

```python
tracks = tracks.dropna()
```

```python
tracks.isnull().sum()
```

```python
tracks.info()
```

```python
sorted_tracks = tracks.sort_values('popularity',ascending = True).head(10)
sorted_tracks
```

```python
tracks.describe().transpose()
```

```python
most_popular = tracks.query('popularity>90',inplace = False).sort_values('popularity',ascending= False)
most_popular[:10]
```

```python
tracks.head(10)
```

```python
tracks[["artist_name"]].iloc[18]
```

```python
tracks["duration"]= tracks["duration_ms"].apply(lambda x: round(x/1000))
tracks.drop("duration_ms", inplace=True, axis=1)
```

```python
tracks.duration.head(10)
```

```python
corr_tracks = tracks.drop(['genre', 'artist_name', 'track_name', 'track_id', 'key', 'mode', 'time_signature'], axis = 1).corr(method='pearson')
plt.figure(figsize=(14, 10))
heatmap = sns.heatmap(corr_tracks, annot=True, fmt=".1g", vmin=-1, vmax=1, center=0, cmap="inferno", linewidths=1, linecolor="black")
heatmap.set_title("Correlation Heatmap Between Variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

```python
sample_track = tracks.sample(int(0.004*len(tracks)))
```

```python
popular_above_50 = tracks[tracks.popularity > 50]
sns.distplot(popular_above_50['acousticness'])
plt.title('Acoustiness for Songs with More than 50 Popularity')
```

```python
popular_below_50 = tracks[tracks.popularity < 50]
sns.distplot(popular_below_50['acousticness'])
plt.title('Acoustiness for Songs with Less than 50 Popularity')
```

```python
sns.distplot(popular_above_50['loudness'])
plt.title('Loudness for Songs with More than 50 Popularity')
```

```python
popular_below_50 = tracks[tracks.popularity < 50]
sns.distplot(popular_below_50['loudness'])
plt.title('Loudness for Songs with Less than 50 Popularity')
```
